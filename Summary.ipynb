>Trong les này ta học ***groupby** method và những thứ liên quan quanh nó.

Nếu muốn get một giá trị đơn cho mỗi group, ta dùng **aggregate()** .

Nếu muốn get một subset của original rows, dùng **filter()**.

Nếu muốn get new value cho mỗi orginal row,dùng **transform()**

```python
import pandas as pd
df= pd.read_csv(
     r"https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"    
)
df.head(10)
```

Nếu muốn tính giá trị trung bình các bill trong một ngày ta làm như sau:

```python
df.groupby('day').agg('mean')
```

Như trên ta thấy những columns nào là kiểu int64 thì nó đều tính **mean** , nếu ta chỉ care 1 cột nào đó ( ở đây t chỉ care total_bill column) thì t làm như sau:

```python
df.groupby('day')['total_bill'].aggregate('mean') # note: agg~ aggregate
# hoặc dùng : df.groupby('day')['total_bill'].mean()
```

Bây giờ ta đặt ra một câu hỏi :

***Làm sao lấy ra những bữa ăn trên những hôm mà average bill >20?***

Ta cần lấy ra những subset của các rows nên ta dùng **filter()**.

```python
df.groupby('day').filter(lambda x: x['total_bill'].mean()>20)
```

## Appendix: pandas.date_range() method.
Refer:https://www.geeksforgeeks.org/python-pandas-date_range-method/


**Syntax**: pandas.date_range(start=None, end=None, periods=None, freq=None, tz=None, normalize=False, name=None, closed=None, **kwargs)

Parameters:
- **start** : Left bound for generating dates.
- **end** : Right bound for generating dates.
- **periods** : Number of periods to generate.
- **freq** : Frequency strings can have multiples, e.g. ‘5H’. See here for a list of frequency aliases.
- **tz** : Time zone name for returning localized DatetimeIndex. By default, the resulting DatetimeIndex is timezone-naive.
- **normalize** : Normalize start/end dates to midnight before generating date range.
- **name** : Name of the resulting DatetimeIndex.
- **closed** : Make the interval closed with respect to the given frequency to the ‘left’, ‘right’, or both sides (None, the default).

Returns: DatetimeIndex



## Appendix: Outliers
Refer:https://ongxuanhong.wordpress.com/2016/01/31/lay-va-lam-sach-du-lieu-xu-ly-du-lieu-ngoai-lai-outliers/

Các điểm outliers có ảnh hưởng đên độ chính xác của các mô hình dự đoán.Nên cần phải phát hiện và xử lý các điểm outliers.
Ví dụ:
Ảnh dưới miêu tả ảnh hưởng của outliers đến quá trình fitting data của mô hình linear regression như nào.



### Các khái niệm liên quan đến outliers


#### 1.Left outlier

Là điểm outlier có giá trị cực tiểu (extreamly low) trong mẫu quan sát.

####  2.Right outliers

Là điểm outlier có giá trị cực đại (extreamly large) trong mẫu quan sát.

#### 3.Representative outlier

Là outlier nằm trong tập dữ liệu.Nó đại diện cho các oulier cùng thể loại khác,được giữ lại để phân tích.
Ví dụ:
khi so sánh kích thước các loại động vật trong thiên nhiên, cá voi đại diện cho động vật lớn nhất, chuột đại diện cho động vật nhỏ nhất.
Giải thích: ở đây cá voi và chuột là representative outlier đc giữ lại để phân tích.

**NOTE**: những điểm presentative outliers này là những điểm được quan sát chính xác( thu thập và ghi chép đúng ).Nếu không chúng trở nên vô nghĩa .

#### 4.Nonrepresentative outlier


Là outlier trong tập data.

Nó chưa được quan sát chính xác.

Được xem là duy nhất trong quần thể vì không tồn tại một giá trị nào tương tự như điểm này.

#### 5.Alpha-trimmed mean


Alpha là giá trị trung bình của tập data. Trong đó alpha/2 trên và dưới của tập data sẽ bị loại bỏ.

####  6.Alpha-winsorized mean


Alpha là giá trị trung bình của tập dữ liệu. Trong đó alpha/2 trên và dưới sẽ được thay thế hoặc chuyển đổi sao cho phù hợp với tập dữ liệu hiện tại.

Ví dụ:

Giá trị x1 (min) và x10( max) sẽ được thay thế bởi giá trị gần chúng nhất là  x2 và x9.

### Way to calculate Outliers
Refer: https://www.wikihow.com/Calculate-Outliers

#### Solution 1:

##### Step1: Lern how to recognize potential outliers.

Ta quan sát và xác định outliers tiềm năng.Nó là thường là các điểm khác biệt lớn với các giá trị khác.ta thường biểu diễn bằng graphs để có thể quan sát được rỡ những điểm outliers.



#####  Step 2:Arrange all data points from lowest to highest.

Ta sắp xếp bộ dữ liệu sau đó tìm median(middle) values của bộ dữ liệu.


#####  Step3: Calculate the median of the data set.

Nếu số lượng phần tử trong data set mà lẻ thì lấy phần tử ở giữa.Còn không lấy trung bình của 2 cái giữa.

##### Step4: Calculate the lower quartile

Là điểm nằm ở 1/4 của bộ dữ liệu đã sắp xếp trên, nếu chắn thì cũng lấy trung bình cộng như trên.

##### Step5: Calculate the upper quartile

Do the same things. nhưng lấy điểm 3/4.


##### Step6: Find the interquartile range

Ta xét hiệu Q3 và Q1 để tìm ra interquartile range để xác định boundary  cho những điểm non- outlier trong data set của ta.
**Note**: Nó làm việc thậm chí Q1,Q3 hoặc cả hai là số âm. Cho ví dụ nếu giá trị Q1 = -70, khi đó interquartile range là 71.5-(-70) = 141.5, cái này là đúng.

##### Step7: Find the "inner fances" for the data set.

-Xác định outliers bằng cách xem xét nó có nằm trong một tập hợp "inner fences" và "outer fences" hay không.

Một điểm mà nằm ngoài inner fences của data set được phân loại như **minor outlier**, trong khi những điểm nằm ngoài out fences được gọi là **major outlier**

***Để tìm inner fences cho data set, đầu tiên ta nhân interquartile range cho 1.5. Sau đó cộng (+) kết quả cho Q3 và trừ (-) kết quả cho Q1. 2 kết quả thu được là  2 boundary của inner fences.***

Ta đã xác định 300 temperature là nằm ngoài range trên nên nó là 1 minor outlier.


##### Step8: Find the "outer fences" for the data set.

Do the same things. Nhưng ta nhân với 3 chư không phải 1.5.Kết quả vẫn +với Q3 và - với Q1.

Những điểm nằm ngoài outter fences được xem là major outliers. Trong ví dụ trên thì 300 nằm ngoài nên nó là major outliers.



##### Step9: Use a qualitative assessment to determine whether to "throw out" outliers.

Xem xét kỹ lần nữa phản đoán xem nó có sát với thực tế k. vì dụ tính ra nhiệt độ trung bình trong phòng gần 90 độ thi rõ là kết quả sai rồi.

```python
import pandas as pd

df =pd.DataFrame({'labels':['a','a','b','b'],'value':[10,20,30,40]})
group = df.groupby('labels')['value'] #gom nhóm theo 'label' và chỉ lấy thêm cột value.
df['value.sum'] = group.transform(lambda x : x.sum()) #hoặc df['value.sum'] = group.transform('sum')
df
```

### How i create a column based on two other columns? 

```python
df = pd.DataFrame({'col1':['minus','minus','positive','nan'],
                  'col2':[10,20,30,40]})
df
```

```python
import numpy as np

df['col3'] = df['col2']*df['col1'].apply(lambda x : -1 if x=='minus'else(1 if x =='positive' else np.nan))
df
```

### How can i apply a function to a group and add the results to my original data frame? 

```python
df = pd.DataFrame({'group1':['a','a','a','b','b','b'],
                  'group2':['c','c','d','d','d','e'],
                  'value1':[1.1,2,3,4,5,6],
                  'value2':[7.1,8,9,10,11,12]})
df
```

```python
group = df.groupby(['group1','group2'])

def Half(x):
    return x.sum()
df['New'] = group['value1'].transform(Half)
df
```

Để thêm nhiều cột( tức áp dụng nhiều nhiều hàm ta dùng **agg()** method.( cái nay đã có Appendix nói rồi ).

```python
#For multiple functions 
def HalfPlus(x):
    return x.sum() +1
newcol = group['value1'].agg([Half,HalfPlus])
newcol
```

## Appendix: Pandas.apply()

>**Pandas.apply** cho phép người dùng pass một function và apply nó trên mỗi single value của **Pandas Series**.Nên nhớ là Series nhé, không phải DataFrame và nhất định cứ không phải là DataFrame.!!!

### Syntax:

**df.apply(func,convert_dtype = True,args=())**

Trong đó:
- func : apply takes a function and applies it to all values of pandas series.
- convert_dtype : Convert dtype as per the function's operation.
- args=() : Additional arguments to pass to function instead of series.
- return type : pandas series after applied function/operation.

### Example: 

```python
import pandas as pd
#reading csv
series = pd.read_csv(r'E:\COMPANY\CSV FILE\stock.csv',squeeze = True)#set 
#squeeze = true :reutruns pandas series if there is only one colunmn 
# define một func để check price.Phải àm điều này vì tý mới dùng dc apply method.
def func(num):
    if num<200:
        return "Low"
    elif num>=200 and num<400:
        return "Normal"
    else:
        return "High"
#passing func to apply and storing returned series in new
new = series.apply(func)
#printing first 3 element
print(new.head(3))
```

## Appendix: Pandas.Merging()/Joining 

Dùng để merge 2 DataFrame lại với nhau thông qua một columns chung.( giống join trong SQL)

**Syntax** :

***pd.merge(left,right,how ='inner',on = None,feft_on = None,right_on = None,
left_index = False,right_index = False,sort = True)***.

Trong đó:
- left: A DataFrame object.
- right: Another DataFrame object.
- on:Columns(names) to join on. Must be found in both the left and right DataFrame object.
- left_on : Columns from the left DataFrame to use as keys.Can either be columns names or arrays with length to the length of the DataFrame.
- right_on : Columns from the right DataFrame to use as keys. Can either be columns names or array with length equal to the length of the DataFrame.
- left_index : If True,use index (row labels) from the left DataFrame as its join key.In case of a DataFrame with a MultiIndex (hierarchical) ,the number of level must match the number of join keys from the right DataFrame.
- right_index: Same usage as left_index for the right DataFrame.
- how : One of 'left','right','outer','inner'. Defaults to innner. Each method has been described below.
- sort: sort the result DataFrame by the join keys in lexicographical order. Defualts to True,setting to False will improve the performance substantially in many cases.


### How to add two data frames and not get null values? 

```python
#create first DF
df1 = pd.DataFrame(data=[26371, 1755, 2], index=[-9999, 240, 138.99], columns=['value'])
df1
```

```python
#create second DF
df2 = pd.DataFrame([26371,1755,6,4],index = [-9999,240,113.03,110],columns = ['value'])
df2
```

```python
#if u simply add them, u will get null values.
# were the index does not match.
df1+df2
```

```python
#Here we fix this issue
df1.add(df2,fill_value = 0)
```
